```python
df = pd.DataFrame(receitas)
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                             content=text,
                             title=title,
                             task_type="RETRIEVAL_DOCUMENT"
                            )["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Receita"], row["Preparo"]), axis=1)
df
```

```python
def gerar_buscar_consulta(consulta, base, model):
    embedding_da_consulta = genai.embed_content(model=model,
                                                content=consulta,
                                                task_type="RETRIEVAL_QUERY"#para realizar buscas
                                               )["embedding"]

    produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

    indice = np.argmax(produtos_escalares)
    return df.iloc[indice]["Preparo"]
```

```python
safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE",
}
```

```python
consulta = input("Qual bolo iremos assar hoje? ")
trecho = gerar_buscar_consulta(consulta, df, model)
generation_config = {"temperature": 0.7,
                     "candidate_count": 1,}
prompt = trecho

model2 = genai.GenerativeModel("gemini-1.0-pro",
                               generation_config=generation_config)
response = model2.generate_content(prompt)
chat = model2.start_chat(history=[])

while prompt != "fim":
  response = chat.send_message(prompt)
  print("Resposta: ",response.text, "\n")
  prompt = input("Me pergunte o que quiser, caso deseje finalizar nossa conversa digite fim: ")
```
